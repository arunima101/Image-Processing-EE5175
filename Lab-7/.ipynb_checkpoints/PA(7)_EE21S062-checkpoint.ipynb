```python
import numpy as np
import scipy
from scipy.fft import fft, ifft, fftshift,ifftshift
import math as m
import random
import matplotlib.pyplot as plt
import cv2
import cmath
```

```python
fourier=cv2.imread('fourier.pgm',0)
fourier_transform=cv2.imread('fourier_transform.pgm',0)
```

```python
def dft_2d(image):
    m,n=np.shape(image)
    
    transform1=np.zeros((m,n),dtype='complex')
    transform_final=np.zeros((m,n),dtype='complex')
    
    
    for i in range(m):
        transform1[i,:]=fft(image[i,:])
        
    for j in range(n):
        transform_final[:,j]=fft(image[:,j])
        
    transform_final=fftshift(transform_final)
    magnitude=np.abs(transform_final)/(m*n)
    
    
    phase=np.zeros_like(transform_final)
    phase[magnitude!=0]=transform_final[magnitude!=0]/magnitude[magnitude!=0]
    
    return magnitude, phase
```

```python
def idft_2d(image):
    m,n=np.shape(image)
    
    transform1=np.zeros((m,n),dtype='complex')
    transform_final=np.zeros((m,n),dtype='complex')
    
    for i in range(m):
        transform1[i,:]=ifft(image[i,:])
    
    for j in range(n):
        transform_final[:,j]=fft(image[:,j])
    
    transform_final=ifftshift(transform_final)
    
    magnitude=np.abs(transform_final)/(m*n)
    magnitude=magnitude[::-1,::-1]
    
    return magnitude
```

```python
out_mag_1,out_phase_1=dft_2d(fourier)
out_mag_2,out_phase_2=dft_2d(fourier_transform)
```

```python
fig = plt.figure(figsize=(15,10))
fig.add_subplot(1, 2, 1)
plt.imshow(fourier,cmap='gray')
fig.add_subplot(1, 2, 2)
plt.imshow(np.log10(out_mag_1),cmap='gray')
```

```python
fig = plt.figure(figsize=(15,10))
fig.add_subplot(1, 2, 1)
plt.imshow(fourier_transform,cmap='gray')
fig.add_subplot(1, 2, 2)
plt.imshow(np.log10(out_mag_2),cmap='gray')
```

```python
image3=out_mag_1*out_phase_2
image4=out_mag_2*out_phase_1

I3=idft_2d(image3)
I4=idft_2d(image4)
```

```python
fig = plt.figure(figsize=(15,10))
fig.add_subplot(1, 2, 1)
plt.imshow(I3,cmap='gray')
fig.add_subplot(1, 2, 2)
plt.imshow(I4,cmap='gray')
```
