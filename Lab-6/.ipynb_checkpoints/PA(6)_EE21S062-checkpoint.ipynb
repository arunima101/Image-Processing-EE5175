```python
import cv2
import numpy as np
import math as m
import random
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
from mpl_toolkits import mplot3d
```

```python
import scipy.io
from scipy.io import loadmat
image_stack=loadmat('stack.mat') #Loading the images as a stack
```

```python
def convolve2D(image,kernel):
    
    if kernel.shape==(1,): #Returns the image if sigma=0
        return image
    
    kernel = np.flipud(np.fliplr(kernel)) #flipping the kernel
    
    kshape=kernel.shape[0]
    xImgShape = image.shape[0]
    yImgShape = image.shape[1]
    
    
    padding=kshape//2 #Calculating extra zeros to pad in the image
    
    xout=xImgShape+2*padding 
    yout=yImgShape+2*padding
    
    
    
    output = np.zeros((xout,yout))
    output[padding:-padding,padding:-padding]=image 
    
    output_image = np.zeros((xImgShape,yImgShape)) 
    
    for x in range(padding,image.shape[0]+padding):
        for y in range(padding,image.shape[1]+padding):
            output_image[x-padding,y-padding]=(kernel * output[x-padding: x + padding+1, y-padding: y + padding+1]).sum()
           
    return output_image
                      
```

```python
def modified_convolve(image,kernel,q,i,j): #This function returns just the convolution of image with kernel
    if q==0:                               #centered around the pixel value given and not with the whole image.       
        return image[i,j]
    
    x,y=np.shape(image)
    
    output = np.zeros((x+2*q,y+2*q)) #q is the padding here
    output[q:-q,q:-q]=image 
    
    m=i+q
    n=j+q
    value=(kernel*output[m-q:m+q+1,n-q:n+q+1]).sum()
    return value
        
```

```python
def ML(image_stack):   
    #This function convolutes the kernel_x and kernel_y and and add the absolute value of them to find 
    #out the modified laplacian, for the entire stack of images.
    
    kernel_x=np.array([[0,0,0],[1,-2,1],[0,0,0]])
    kernel_y=np.array([[0,1,0],[0,-2,0],[0,1,0]])
    modified_stack={}
    
    # 100 images are given
    
    for i in range(1,101):
        if i<10:
            imagex=convolve2D(image_stack['frame00'+str(i)],kernel_x)
            imagey=convolve2D(image_stack['frame00'+str(i)],kernel_y)
            modified_image=np.abs(imagex)+np.abs(imagey)
            modified_stack.update({'frame00'+str(i):modified_image})
        if i>=10 and i<100:
            imagex=convolve2D(image_stack['frame0'+str(i)],kernel_x)
            imagey=convolve2D(image_stack['frame0'+str(i)],kernel_y)
            modified_image=np.abs(imagex)+np.abs(imagey)
            modified_stack.update({'frame0'+str(i):modified_image})
        else:
            imagex=convolve2D(image_stack['frame100'],kernel_x)
            imagey=convolve2D(image_stack['frame100'],kernel_y)
            modified_image=np.abs(imagex)+np.abs(imagey)
            modified_stack.update({'frame100':modified_image})
    
    return  modified_stack    
            
            
```

```python
def depthmap(modified_stack,q):  #Takes, Modified laplacian stack of images and sum it up around 
                                 #a neighbourhood of N=2*q+1. The sum is stored in focus measure,
                                 #and the dbar corresponding to max focus measure is recorded in depth
    N=2*q+1                      #map. 
    q_kernel=np.ones((N,N))
    x,y=np.shape(modified_stack['frame001'])
    
    depth=[]
    
    for i in range(x): 
        for j in range(y):
            focus_measure=[]
            dbar=[]
            for k in range(1,101):
                if k<10:
                    val_ret=modified_convolve(modified_stack['frame00'+str(k)],q_kernel,q,i,j)
                    focus_measure.append(val_ret)
                    dbar.append((k-1)*50.5)
                if k>=10 and k<100:
                    val_ret=modified_convolve(modified_stack['frame0'+str(k)],q_kernel,q,i,j)
                    focus_measure.append(val_ret)
                    dbar.append((k-1)*50.5)
                else:
                    val_ret=modified_convolve(modified_stack['frame100'],q_kernel,q,i,j)
                    focus_measure.append(val_ret)
                    dbar.append((k-1)*50.5)
                    
            val=focus_measure.index(max(focus_measure))
            depth.append(dbar[val])
        
            
      
    return depth   
        
    
```

```python
a=ML(image_stack)
```

```python
depth0=depthmap(a,0)
depth_array_0=np.reshape(depth0,(115,115))
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.gca(projection='3d')
x, y = 115,115
X, Y = np.meshgrid(np.arange(x)+1, np.arange(y)+1)
surf = ax.plot_surface(
    X, Y, depth_array_0, cmap="Blues", linewidth=0, antialiased=False)

plt.title("For q=0")
fig.colorbar(surf)
plt.show()
```

```python
depth1=depthmap(a,1)
depth_array_1=np.reshape(depth1,(115,115))
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.gca(projection='3d')
x, y = 115,115
X, Y = np.meshgrid(np.arange(x)+1, np.arange(y)+1)
surf = ax.plot_surface(
    X, Y, depth_array_1, cmap="Purples", linewidth=0, antialiased=False)

plt.title("For q=1")
fig.colorbar(surf)
plt.show()
```
